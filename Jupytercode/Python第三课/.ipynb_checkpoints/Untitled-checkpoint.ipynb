## 函数封装 获取列表下面所有租房信息链接

```python
def get_links(url):
    responce= requests.get(url)
    soup = BeautifulSoup(responce.text, 'lxml')
    links_div = soup.find_all('div',class_="zu-itemmod")
    links = [div.a.get('href') for div in links_div]
    return links
```

# 完整爬取加存储数据

```python
import requests
from bs4 import BeautifulSoup 
import pymysql
import time
```

```python
def get_pages(url):
    responce= requests.get(url)
    soup = BeautifulSoup(responce.text, 'lxml')
    return soup
def get_links(link_url):
    soup = get_pages(link_url)
    links_div = soup.find_all('div',class_="f-list-item ershoufang-list")
    links = [div.a.get('href') for div in links_div]
    return links
```

```python
link_url='http://bj.ganji.com/zufang/'
links_lj=get_links(link_url)
```

```python
#links_lj 网页所以有链接地址
links_lj,len(links_lj)
```

# 爬取具体页面信息

```python
def get_house_iof(ho_url):
    soup=get_pages(ho_url)
    title=soup.find('p',class_="card-title").text.strip()
    price=soup.find('span',class_="price").text
    house_in = soup.find_all('span',class_="content")
    stytle=house_in[0].text
    area=house_in[1].text[4:]
    direction=house_in[2].text
    floor=house_in[3].text
    conditions=house_in[4].text
    name_n=house_in[5].text[2:6]
    subway=house_in[6].text
    location=house_in[7].text.strip()
    agent=soup.find('a',class_="name").text
    card=soup.find('div',class_="work_license_box").text[6:13]
    info = {
        '标题描述':title,
        '价格':price,
        '户型':stytle,
        '面积':area,
        '朝向':direction,
        '楼层':floor,
        '装修情况':conditions,
        '小区名称':name_n,
        '地铁':subway,
        '位置':location,
        '经理人':agent,
        '从业信息卡':card
    }
    return info
```

```python
ho_url = 'http://bj.ganji.com/zufang/38704376726535x.shtml?entinfo=38704376726535_0&fzbref=0&params=rank0830gspanxuan0099^desc&gjcity=bj&psid=174787436204870949158019915&cookie=|||18104ad0828a3a9db2bd79c7a2df031d&apptype=10&key=&pubid=81235422&trackkey=38704376726535_c831b2c3-4f5f-426b-af8d-63a65adb4b53_20190714021921_1563041961541&fcinfotype=gz&jingxuan=1'
info = get_house_iof(ho_url)
info
```

# 存贮到数据库

```python
DATABASE = {
    'host': '127.0.0.1',
    'database': 'house_iof',
    'user': 'root',
    'password': 'lwzdxydz2'
}

def get_db(setting):
    return pymysql.connect(**setting)

def insert(db,inf):
    cursor = db.cursor()
    values_1 = "'{}',"*11+"'{}'"
    sql_values = values_1.format(inf['标题描述'],inf['价格'],inf['户型'],inf['面积'],inf['朝向'],inf['楼层'],inf['装修情况'],inf['小区名称'],inf['地铁'],inf['位置'],inf['经理人'],inf['从业信息卡'])
    sql = """INSERT INTO house(title,price,stytle,area,direction,floor,conditions,name_n,subway,location,agent,card) 
             VALUES({});""".format(sql_values)
    cursor.execute(sql)
    db.commit()
```

```python
links_2j = links_lj[0:3]
i=0
db = get_db(DATABASE)
for ho_url in links_2j:
    time.sleep(5)
    info = get_house_iof(ho_url)
    print(i,':',info)
    i+=1
    insert(db,info)
```
